```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
iris_df = pd.read_csv("/Users/aryantele/Downloads/Iris.csv")
iris_df.drop(columns='Id', inplace=True,)
```

```python
iris_df.info()
```

```python
# Plot histograms for each feature
iris_df.hist(figsize=(10, 8))
plt.suptitle('Histograms of Features in Iris Dataset')
plt.show()
```

```python
# Plot boxplots for each feature
iris_df.boxplot(figsize=(10, 8))
plt.title('Boxplots of Features in Iris Dataset')
plt.show()
```
