```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/aryantele/Downloads/archive (1)/covid_vaccine_statewise.csv")
```

```python
df.head(20)
```

```python
df2 = df[['State', 'Total Doses Administered', 'First Dose Administered', 'Second Dose Administered','Male (Doses Administered)', 'Female (Doses Administered)']]
```

```python
df2.head()
```

```python
df2['Male (Doses Administered)'].fillna(np.mean(df['Male (Doses Administered)']), inplace=True)
```

```python
df2['Female (Doses Administered)'].fillna(np.mean(df['Female (Doses Administered)']), inplace=True)
```

```python
df2['First Dose Administered'].fillna(np.mean(df['First Dose Administered']), inplace=True)
```

```python
df2['Second Dose Administered'].fillna(np.mean(df['Second Dose Administered']), inplace=True)
```

```python
df2['Total Doses Administered'].fillna(np.mean(df['Total Doses Administered']), inplace=True)
```

```python
df2.isna().sum()
```

```python
df2.info()
```

```python
df2['State'] = df2['State'].astype('category')
```

```python
df2.info()
```

```python
df2.describe()
```

```python
df2.shape
```

```python
plt.figure(figsize=(70,60))
sns.histplot( x="State",y="First Dose Administered", data=df2)
```

```python
df2 = df2[df2['State'] != 'India']
```

```python
df2 = df2[df2['First Dose Administered'] != 0]
```

```python
df2.to_csv('Cleaned_Covied19.csv')
```
