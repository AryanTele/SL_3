```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./datasets/titanic_cleaned.csv')
df.drop(columns='Unnamed: 0', inplace= True, axis=1)
```

```python
df.head()
```

```python
sns.countplot(x='Survived', data=df)
```

```python
sns.countplot(x='Survived',hue='Sex', data=df)
```

```python
sns.countplot(data=df, x="Survived", hue=df['Pclass'].astype(str))
```

```python
corr = df[['Age', 'Pclass','Survived', 'Fare']].corr()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(corr, annot=True)
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=df, x='Pclass', y='Age')
```

```python
plt.figure(figsize=(20,10))
sns.histplot(data=df, x='Fare',hue='Survived',kde=True)
```

```python
plt.figure(figsize=(20,10))
sns.histplot(data=df, x='Fare',hue='Sex',kde=True)
```
